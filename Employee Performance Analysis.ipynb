```python
y = data["Performance_Score"]
x = data[["Years_At_Company","Monthly_Salary","Overtime_Hours","Promotions","Employee_Satisfaction_Score"]]
```

```python
x
```

```python
x.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
#scaled x train
x_train = scaler.fit_transform(x_train)
```

```python
import joblib
joblib.dump(scaler,"scaler.pkl")
```

```python
#scaled x test
x_test = scaler.fit_transform(x_test)
```

```python
x_test
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.metrics import accuracy_score
```

```python
def modelperformance(predictions):
    print("Accuracy Score on model is{}".format(accuracy_score(y_test, predictions)))
    
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(x_train, y_train)
```

```python
log_model.predict([[1, 1, 1, 1, 1]])
```

```python
log_model.predict(x_test)
```

```python
log_predictions = log_model.predict(x_test)
```

```python
modelperformance(log_predictions)
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors":[3, 5, 7, 9, 11], "weights":["uniform", "distance"]}
```

```python
gridkn = GridSearchCV(KNeighborsClassifier(), param_grid, cv = 5)
```

```python
gridkn.fit(x_train, y_train)
```

```python
gridkn.best_params_
```

```python
prediictions = gridkn.predict(x_test)
```

```python
modelperformance(prediictions)
```
